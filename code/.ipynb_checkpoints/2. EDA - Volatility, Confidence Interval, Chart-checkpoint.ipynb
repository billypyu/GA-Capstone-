**Distributions and Confidence Intervals**

We establish the confidence intervals to measure uncertainty. A higher probability associated with the confidence interval means that there is a greater degree of certainty that the parameter falls within the bounds of the interval. Therefore, a higher confidence level indicates that the parameters must be broader to ensure that level of confidence. Also, low stdev is closed to the mean. High stdev means data is dispreaded over a wide range of value.

```python
def conf_interval(the_sample):
    '''build a 95% conf int for analysis'''
    the_mean = np.mean(the_sample)
    the_st = np.std(the_sample)
    n = len(the_sample)
    margin = 1.96 * the_st / np.sqrt(n)
    return (the_mean - margin, the_mean + margin)

print(f'95% Confidence Intervals on ESP: {conf_interval(df.esp)}')
print(f'Mean: {df.esp.mean()}')
print(f'Median: {df.esp.median()}')
print(f'Mode: {df.esp.mode()}')
```
